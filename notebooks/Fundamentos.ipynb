# O que é LangChain e para que Serve ?

**LangChain** é uma biblioteca que facilita a construção de aplicações de prcessamento de linguagem natural (**NLP**) utilizando grandes modelos de linguagem (**LLMs**). Ela permite a integração de LLMs com ferramentas e workflows, simplificando o desenvolvimento de aplicações complexas como chatbots, análise de texto e geração de conteúdo.

Nessa primeira sessão não vamos usar o LangChain ainda. Primeiro vamos entender os tipos de modelos ( Completion e Chat ) e usaremos a **OpenAI** com os métodos da OpenAI.

# Modelos OpenAI

## Modelos de Completion (Legados)
Os modelos de Completion foram o padrão original (como o GPT-3 ``davinci``). Eles funcionam de forma linear: você fornece um texto inicial (**prompt**) e o modelo gera uma continuação que parece estatisticamente provável.
- **Lógica de Funcionamento**: "Dado este texto, o que vem a seguir?"
- **Formato de Entrada**: Uma única String (texto puro).
- **Exemplo**:
    - **Prompt**: "A capital da França é"
    - **Output**: " Paris, e é conhecida como a Cidade Luz."
## Modelos de Chat (Atuias)
Os modelos de Chat (como o Gemini ou GPT-4) são otimizados para diálogo. Em vez de apenas prever a próxima palavra de um texto solto, eles são treinados para entender turnos de fala e contextos de conversação. 
- **Lógica**: "Como devo responder a esta sequência de mensagens?"
- **Formato de Entrada**: Uma **Lista de Objetos** , onde cada objeto tem um **papel** (_role_) e um conteúdo.
- **Papéis Principais**:
    - **System**: Define o comportamento e as regras do assistente (ex: "Você é um professor de Python").
    - **User**: A mensagem enviada por você.
    - **Assistant**: As respostas anteriores da IA (usadas para manter o contexto).
